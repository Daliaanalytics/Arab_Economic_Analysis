```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
arab_countries= ['Saudi Arabia', 'United Arab Emirates', 'Egypt, Arab Rep.',
                  'Jordan', 'Morocco', 'Tunisia']
df_arab= df[df['country_name'].isin(arab_countries)]
```

```python
cols_to_keep= ['country_name', 'year',
                'GDP (Current USD)',
                'GDP per Capita (Current USD)',
                'GDP Growth (% Annual)',
                'Inflation (CPI %)',
                'Unemployment Rate (%)',
                'Gross National Income (USD)']
df_arab= df_arab[cols_to_keep]
```

```python
df_arab.shape
```

```python
df_arab.isnull().sum()
```

```python
df_arab_backup= df_arab.copy()
```

```python
df_arab.head()
```

```python
numeric_cols= ['GDP (Current USD)', 'GDP per Capita (Current USD)',
                'GDP Growth (% Annual)', 'Inflation (CPI %)',
                'Unemployment Rate (%)', 'Gross National Income (USD)']
df_arab[numeric_cols]= df_arab.groupby('country_name')[numeric_cols].transform(lambda x: x.fillna(x.mean()))
```

```python
df_arab.isnull().sum()
```

```python
df_arab.head(10)
```

```python
df_arab.shape
```

```python
df_arab['GDP (Billion USD)']= df_arab['GDP (Current USD)']/1e9
```

```python
df_arab['GNI (Billion USD)']=df_arab['Gross National Income (USD)']/1e9
```

```python
df_arab[['country_name', 'year', 'GDP (Billion USD)']].head()
```

```python
df_arab.groupby('country_name')['GDP (Billion USD)'].mean().sort_values(ascending= False)
```

```python
Avg_GDP= df_arab.groupby('country_name')['GDP (Billion USD)'].mean().sort_values(ascending= False)
plt.bar(Avg_GDP.index, Avg_GDP.values)
plt.title('Average of GDP')
plt.xlabel('country_name')
plt.ylabel('GDP (Billion USD)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,6))

for country in arab_countries:
  data= df_arab[df_arab['country_name']== country]
  plt.plot(data['year'], data['GDP (Billion USD)'], label=country, marker='o')

plt.title('GDP Growth Over Time')
plt.xlabel('year')
plt.ylabel('GDP (Billion USD)')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
numeric_cols= ['GDP (Billion USD)', 'GDP per Capita (Current USD)',
                'GDP Growth (% Annual)', 'Inflation (CPI %)',
                'Unemployment Rate (%)', 'GNI (Billion USD)']
correlation_matrix= df_arab[numeric_cols].corr()
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, fmt= '.2f', cmap='coolwarm')
plt.title('Correlation Heatmap - Arab Countries')
plt.show()
```
